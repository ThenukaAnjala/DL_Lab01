Create a random 100000x1 array from the same exponential distribution and visualize the
distribution histogram along with uniform and normal distributions.

```python
# 100000x1 array and histogram
exp_array = np.random.exponential(scale=1.0, size=100000)
uniform_array = np.random.rand(100000)
normal_array = np.random.randn(100000)

plt.hist(exp_array, density=True, bins=100, histtype='step', color='red', label='exponential')
plt.hist(uniform_array, density=True, bins=100, histtype='step', color='blue', label='uniform')
plt.hist(normal_array, density=True, bins=100, histtype='step', color='green', label='normal')

plt.axis([-2.5, 5, 0, 1.1])
plt.legend(loc='upper right')
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot.

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
